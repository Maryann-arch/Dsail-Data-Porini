```python
#creating a column for time 
df['Time']=df['Filename'].str.extract('(....-..-..-..-..-..)',expand=False)
df = df.dropna()
```

```python
df
```

```python
#Converting the datetime column into datetim_objects
df['datetime_object'] =  pd.to_datetime(df['Time'], format='%Y-%m-%d-%H-%M-%S')
df
```

```python
#SORTING THE DATA ACCORDING TO DATES
df.sort_values(by='datetime_object', inplace=True, ignore_index=True)
df
```

```python
datetime_lists = []

delta_minutes = 10 #steps to increase the time interval in minutes
t1 = df.iloc[0].datetime_object


while t1 < df.iloc[-1].datetime_object:
    datetime_list = []
    datetime_list.extend([t1, t1 + pd.DateOffset(minutes=delta_minutes)])
    t1 += pd.DateOffset(minutes=delta_minutes)
    datetime_lists.append(datetime_list)
    
datetime_lists[-1]
```

```python
len(datetime_lists)
```

```python
#creating lists of the different variables using a time interval of 10 minutes using the time and species

times_list = []
species_list = []
filtered_df = pd.DataFrame()
for t_range in datetime_lists:
    dt=[]
    sp_list = []
    for i in range(len(df)):
        item = df['datetime_object'][i]
        species = df['Species'][i]
        if item >= t_range[0] and item <t_range[1]:
            # do something hereS
            dt.append(item)
            sp_list .append(species)
            
    if len(dt) > 0:     
        times_list.append(dt)
        species_list.append(sp_list)
```

```python
camera_trap = pd.DataFrame(columns=['species', 'times'])
for i in range(len(species_list)):
    result = all(word == species_list[i][0] for word in species_list[i])
    if result:
                #print(species_list[i][0], '\t', np.random.choice(times_list[i]))
                #print(np.random.choice(times_list[i]))
                camera_trap = camera_trap.append({'species':species_list[i][0], 'times':np.random.choice(times_list[i])}, ignore_index=True)

    else:
            sps = sorted(set(species_list[i]))
            ts = list(np.random.choice(times_list[i], len(sps)))
            #print(sps, ts)
            for indx, sp in enumerate(sps):
                #print(sp, ts[indx])
                camera_trap.append([sp, ts[indx]])
                camera_trap = camera_trap.append({'species':species_list[i][0], 'times':np.random.choice(times_list[i])}, ignore_index=True)

        
#pd.dataframe(d)
```
